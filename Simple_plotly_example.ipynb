```python
import plotly as py
import plotly.graph_objs as go
import numpy as np
```

```python
x= np.linspace(0,np.pi,100)
y = np.sin(x)
```

```python
layout = go.Layout(
title = 'Simple graph',
yaxis = dict(title = 'sin'),
xaxis = dict(title='x')
)
```

```python
trace1 = go.Scatter(
x=x,
y=y,
mode='markers',
name = 'sinx',
marker= dict(color='red',size=[10]*1000))
```

```python
fig = go.Figure(data=[trace1],layout=layout)
```

```python
py.offline.init_notebook_mode(connected=True)
py.offline.iplot(fig)
```
